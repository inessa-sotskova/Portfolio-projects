# Исследование оттока клиентов в сети фитнес-клубов #

## Цель и задачи проекта ##

**Цель:** провести анализ анкет клиентов и подготовить план действий по удержанию клиентов.  

**Задачи:**
- спрогнозировать __вероятность оттока__ (на уровне следующего месяца) для каждого клиента;
- сформировать __типичные портреты клиентов__: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать __признаки, наиболее сильно влияющие на отток__;
- сформулировать основные выводы и __разработать рекомендации__ по повышению качества работы с клиентами:

    1) выделить целевые группы клиентов;
    2) предложить меры по снижению оттока;
    3) определить другие особенности взаимодействия с клиентами.


## Описание данных ##

__Данные клиента за предыдущий до проверки факта оттока месяц:__

- `gender` — пол;
- `Near_Location` — проживание или работа в районе, где находится фитнес-центр;
- `Partner` — сотрудник компании – партнёра клуба;
- `Promo_friends` — факт первоначальной записи по промо-коду в рамках акции «приведи друга»;
- `Phone` — наличие контактного телефона;
- `Age` — возраст;
- `Lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).


__Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:__
- `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
- `Group_visits` — факт посещения групповых занятий;
- `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
- `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
- `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
- `Churn` — факт оттока в текущем месяце.

## 3. Построение модели прогнозирования оттока клиентов ##

### 3.1 Разделение датасета на признаки и целевую переменную; обучающую и валидационную выборки. Стандартизация признаков. Выбор метрик. ###

Разделим датасет на признаки (матрица X) и целевую переменную "отток" (y)

```python
X = df.drop('churn', axis = 1)
y = df['churn']
```

Разделим модель на две на обучающую и валидационную выборки. Чтобы для обучения модели можно было использовать линейную регрессию, проведем стандартизацию признаков. Создадим объект класса StandardScaler и применим его к обучающей выборке и к к матрице признаков для тестовой выборки:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0, stratify = y)

scaler = StandardScaler() 
X_train_st = scaler.fit_transform(X_train) 
X_test_st = scaler.transform(X_test)
```

<div class="alert alert-info">
    <b>Комментарий студента:</b> Спасибо, применила, и метрики заметно улучшились. И в общем-то, не так нужно стало перебирать пороги) Но я оставлю пункт 3.4 для себя, чтобы не забыть, как это делается
</div>

Зададим функцию для вывода метрик, которые помогут оценить качество моделей

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

### 3.2 Модель логистической регрессии для прогнозирования оттока в следующем месяце ###

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression(random_state=0)
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
# выведем все метрики
print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии:',
)
```

### 3.3 Модель на основе алгоритма случайного леса для прогнозирования оттока в следующем месяце ###

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state=0)

rf_model.fit(X_train_st, y_train)

rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]

print_all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities,
    title='Метрики для модели случайного леса:'
)
```

### 3.4 Метрики моделей с измененным трешхолдом ###

Попробуем улучшить качество моделей. Перепробовав трешхолды 0.3, 0.4, 0.45, 0.55, 0.6, определенные улучшения модели показывают с трешхолдом, равном 0.4:
- для модели логистической регрессии ухудшился показатель точности (-0,03), но увеличились показатели полноты (+0,06) и баланса F1 (+0,02), другие показатели остались неизменными; 
- для модели на алгоритме случайного леса также ухудшилась точность (-0,03), но увеличилась полнота (+0,09), метрика баланса F1 (+0,03) и ROC AUC (+0,02), доля правильных ответов осталась неизменной.

```python
# зададим порог
threshold = 0.4

# на основании вероятностей и соотношения классов рассчитаем predict
custom_predictions = [0 if i < threshold else 1 for i in lr_probabilities]
print_all_metrics(
    y_test,
    custom_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии c трешхолдом, равном 0,4:',
)
```

### 3.5 Сравнение моделей ###

 __Обе модели работают достаточно хорошо__. Разберем каждый показатель отдельно (для моделей с автоматическим трешхолдом, равном 0,5):

- __Доля правильных ответов (accuracy)__. Модель LR правильно предугадывает статус клиента (как "0" - остался, так и "1" - ушел в отток) в 90 случаях из ста, а модель RF – в 91 случаях из 100.

- __Точность (precision)__. Показывает долю правильных ответов только среди целевого класса "1" (ушедшие в отток). Здесь модель LR чуть чаще принимала за ушедших в отток тех, кто на самом деле остался (0.83 против 0.86 у модели RF). В нашем случае ложноположительные ответы не так страшны: допустим, модель неправильно предскажет, что человек скоро покинет фитнес-центр. Вряд ли в попытках удержать его фитнес-центр потратит много ресурсов.

- __Полнота (recall)__. Показывает, сколько реальных объектов "1" класса удалось обнаружить с помощью модели. Чем меньше ложнонегативных ответов, тем выше качество модели. В нашем случае LR сработала намного лучше (0.82 против 0.79). В ситуации с фитнес-центром эта метрика важнее, чем точность: при низком показателе полноты мы рискуем не обнаружить вовремя уходящих клиентов и потерять их.
- __F1__ - балансовая мера между точностью и полнотой – одинакова для обеих моделей (0.82)
- __ROC AUC__ – площадь под кривой ошибок чуть больше для модели LR, чем RF (0.96 против 0.95)

В пункте 3.4 мы экспериментировали с качеством моделей с помощью трешхолда. Самый сбалансированный результат – все-таки при пороге 0,5. Например, при пороге 0,4 полнота становится (+ 0,04 для LR и + 0.08 для RF), но при этом  сопоставимо уменьшается точность(- 0,04 для LR и - 0.08 для RF). Поскольку для __прогнозирования оттока полнота (recall) более релевантна, мы скорее порекомендуем модель логистической регрессии с порогом от 0,5 до 0,4__.


## 4. Кластеризация клиентов ##

### 4.1 Кластеризация с помощью дендрограммы ###

Уберем столбец с оттоком и выполним стандартизацию матрицы, передадим ее переменной linked, которая сохранить "связки" между объектами. 

```python
X_full = df.drop('churn', axis = 1)
sc = StandardScaler()
X_sc = scaler.fit_transform(X_full)

linked = linkage(X_sc, method = 'ward') 
```

Визуализируем таблицу связок linked в виде дендрограммы.

Предложенное дендрограммой оптимальное число кластеров равно 4. Визуально же можно выделить также 5 (если разделить зеленый кластер) или 6 (если разделить и зеленый, и фиолетовый) кластеров. 

```python
fig, ax = plt.subplots(figsize=(12, 8))

dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация для клиентов фитнес-клуба')
ax.axhline(y = 44, color = 'grey', linestyle = '--')
ax.axhline(y = 53, color = 'grey', linestyle = '--')
plt.show() 
```

### 4.2 Кластеризация с помощью алгоритма K-Means ###

Обучим модель кластеризации на основании алгоритма K-Means с пятью кластерами и спрогнозируйте кластеры для всех клиентов. В исходном датасете df создадим новый столбец, где каждому клиенту будет присвоен свой номер кластера.

```python
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(X_sc)
```

```python
df['cluster_km'] = labels
```

```python
df.head(3) #проверим наличие столбца с кластерами
```

Для удобства в визуализации данных зададим строковые значения для кластеров – от кластера 1 до кластера 5.

```python
def categorize_clusters(cluster):
    if cluster == 0:
        return 'cluster 1'
    if cluster == 1:
        return 'cluster 2'
    if cluster == 2:
        return 'cluster 3'
    if cluster == 3:
        return 'cluster 4'
    if cluster == 4:
        return 'cluster 5'
```

```python
df['cluster_name'] = df['cluster_km'].apply(categorize_clusters)
```

### 4.3 Описание кластеров ###

__Выведем средние значения признаков__ для кластеров. В таблице видно, что __объединяет кластеры только доля мужчин и женщин, а в остальном есть различия – как небольшие, так и очень заметные__ (напр., полное отсутствие контактных номеров у кластера 3 или почти нулевой отток в кластере 1). Эти и другие наблюдения перенесем в __портреты клиентов кластеров (ниже)__.

```python
df.groupby('cluster_name').mean().T.round(2)
```

__Построим столбчатые диаграммы и диаграммы размаха__ для лучшего понимания распределения признаков в разных кластерах.

```python
for col in df.drop(['cluster_name', 'cluster_km', 'avg_class_frequency_total', 'avg_additional_charges_total'], axis = 1).columns:
    fig, ax = plt.subplots(figsize=(8, 3))
    sns.countplot(x=df[col],
                 hue = df['cluster_name'],
                 palette = 'pastel',
                 hue_order = ['cluster 1', 'cluster 2', 'cluster 3', 'cluster 4', 'cluster 5'])
    
    ax.set_title('Distribution of {} among clusters'.format(col), pad = 10)
    plt.show()
```

```python
for col in df[['avg_class_frequency_total', 'avg_additional_charges_total']].columns:
    fig, ax = plt.subplots(figsize=(12, 4))
    sns.boxplot(x=df[col],
                 y = df['cluster_name'],
                 palette = 'pastel')
    
    ax.set_title('Distribution of {} among clusters'.format(col), pad = 10)
    plt.show()
```

__Рассчитаем показатель оттока в каждом из кластеров__, а также размер и долю каждого кластера.

```python
clusters = (
    df.groupby('cluster_name')
    .agg({'age': 'count', 'churn': 'mean'})
    .round(2)
    .rename(columns = {'age': 'count', 'churn': 'churn_rate'}).
    sort_values('churn_rate', ascending = False)
)
clusters['%_sample '] = (clusters['count'] / len(df) * 100).round(1)
clusters
```

Несмотря на прекрасные пререквизиты (клиенты живут или работают рядом, приходят от друзей и коллег), посетители не задерживаются в клубе. В рекомендациях мы не будем затрагивать качество услуг, оборудования и пр. (хотя это база и крайне важно), но сфокусируемся на установлении долгосрочного сотрудничества с клиентами. Выделим несколько направлений:
- __Забота о клиенте и внимание к нему__:
    - предложить клиентам с отсутствующими номерами оставить их для оперативной связи и спецпредложений;
    - сделать предложение для детей, подростков и старшего поколения. В особенности последние более стабильны и дисциплинированны, а также объективно нуждаются в укреплении здоровья.
    - сделать фокус на огромном кластере 5 (и клиентах из других кластеров с похожими ситуациями): все они живут или работают рядом, но большинство приходят в клуб одни. Им нужен максимальный онбординг. Так, им уже на входе предлагать самые подходящие по здоровью и темпераменту программы, разъяснять про загруженные часы, запрашивать обратную связь. Напомним, в кластере 5 больше всего молодых людей, для кого-то данный центр может быть первым фитнес-клубом в жизни;
    - провести эксперимент с увеличением минимального абонемента до 2 мес (без увеличения цены) для новых клиентов и посмотреть, поможет ли это конвертировать клиентов в более долгосрочные абонементы (за два месяца более вероятно пообвыкнуться, чем за один)
- __Развитие корпоративного направления__:
    - половина клиентов работает в компаниях-партнерах. Можно предлагать внутри- и межкорпоративные спортивные челленджи и соревнования. Это может увеличить приток клиентов и увеличить вовлеченность.
- __Развитие локальной спортивной жизни__:
    - для тех, кто живет рядом, будет актуальным участие в различных мероприятиях, организованных клубом или городом. Поскольку это сеть, можно соревноваться между клубами или устраивать оздоровительные мероприятия или игры на открытом воздухе (зачастую, затраты на это – лишь часы организаторов). Это помогает людям познакомиться друг с другом, с большей охотой заниматься самим и приглашать друзей и знакомых. В качестве лидеров команд и организаторов можно приглашать людей из кластера 1 с большим лайфтаймом.
- __Использовать нашу модель__ для заблаговременного обнаружения тех, кто готов уйти и принять меры по их возвращению.
